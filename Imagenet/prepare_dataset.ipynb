```python
import shutil,os
```

```python
def txt2list(file_path):
    with open(file_path,"r") as f:
        txt_file = f.readlines()

    temp_list = []

    map(lambda x: temp_list.append(x.split("\t")),txt_file)

    return temp_list    

def txt2dict(file_path):
    with open(file_path,"r") as f:
        txt_file = f.readlines()

    temp_dict = {}

    for dirty_line in txt_file:
        
        dirty_line = dirty_line.split("\t")

        key = dirty_line[0]
        context = dirty_line[1:]

        temp_dict[key] = context 
    
    return temp_dict


def getMyClasses(words_file,wnids_file):
   

    classes_names = txt2dict(words_file)
    
    all_classes = classes_names.keys()
    
    my_classes = list(map(lambda x: x.strip(),wnids_file))
    
    for class_name in set(all_classes).symmetric_difference(my_classes):
        
        del classes_names[class_name]

    print("Class names did take...")

    return classes_names

def change_image_files_path():
    
        if os.path.isdir(project_path+"dataset") == False:
        
            os.mkdir(project_path+"dataset")

            
        
        
        for file_name in os.listdir(dataset_path):

            
            
            if file_name.endswith(".txt") :
                
                shutil.copy(dataset_path + file_name,project_path+"dataset/"+file_name)
            

            if file_name=="train" and  os.path.isdir(project_path+"dataset/train") == False:
                
                os.mkdir(project_path+"dataset/"+file_name)
                
                for folder_name in os.listdir(dataset_path+file_name):
                    
                    os.mkdir(project_path+"dataset/"+file_name+"/"+folder_name)
                    
                    for image_file in os.listdir(dataset_path+file_name+"/"+folder_name+"/images"):
                      
                        shutil.copy(dataset_path+file_name+"/"+folder_name+"/images/"+image_file,project_path +"dataset/"+file_name+"/"+folder_name+"/"+ image_file)
            
            
            if file_name =="val" and os.path.isdir(project_path+"dataset/val") == False:
                
                image_categories = txt2dict(dataset_path+file_name+"/val_annotations.txt")

                os.mkdir(project_path+"dataset/"+file_name)
                
                
                

                for image_file in os.listdir(dataset_path+file_name+"/images"):

                    folder_name = image_categories[image_file][0]
                    os.mkdir(project_path +"dataset/"+file_name+"/"+folder_name) if os.path.isdir(project_path +"dataset/"+file_name+"/"+folder_name) == False else print()
                    
                    shutil.copy(dataset_path+file_name+"/"+"/images/"+image_file,project_path +"dataset/"+file_name+"/"+folder_name+"/"+ image_file)
     
        print("Image path changed... ")



def make_dataset_set():
    
    os.chdir(project_path)

    change_image_files_path()
    
   


make_dataset_set()
```
